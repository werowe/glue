```python
import pandas as pd
import numpy as np
import re


 
f='/Users/walkerrowe/Downloads/alpha.csv'

df = pd.read_csv(f,skiprows=2,usecols=[2,5,12,14],thousands=',')

df.columns =['date', 'payee', 'expense', 'deposit']
df.fillna(0,inplace=True)
 
df['Amount']=df['expense'] 

euro = 1.01


dt=pd.Series(df['date'].astype(str).map(lambda l : l[3:5] + '/' + l[0:2] + '/' + l[6:10] ),name="Date")
 
payee=pd.Series(df['payee'],name="Payee")
amount=pd.Series(df['Amount'].astype('float')) * -euro


simplifi=pd.DataFrame([dt,payee,amount.round(2)]).transpose()

simplifi.to_csv(f+'alpha.csv',sep=",",line_terminator='\n',index=False)
```
